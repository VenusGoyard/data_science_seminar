how to plot a dataset:

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.plot(tips['total_bill'],tips['tip'], 'o', alpha = 0.3)
plt.title('Correlation Between Total Bill and Tips')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.grid()
```

```python
plt.hist(tips['total_bill'])
```

```python
plt.boxplot(tips['total_bill'])
```

```python
plt.boxplot(tips['total_bill'], vert = False)
```

```python
sns.countplot(tips['sex'])
```

```python
tips.head()
```

```python
tips['pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
plt.scatter(tips['total_bill'], tips['tip'])
```

```python
sns.scatterplot(tips['total_bill'], tips['tip'], c = tips['pct'], cmap = 'PiYG')
```

```python
sns.scatterplot(tips['total_bill'], tips['tip'], 
                s = tips['pct']*500, c = tips['size']*100, cmap = 'PiYG')
```

splitting by a yes/no categories

```python
sns.boxplot(tips['day'], tips['total_bill'], hue = tips['smoker'])
```

```python
sns.boxplot(tips['day'], tips['pct'], hue = tips['sex'])
```

```python
penguins = sns.load_dataset('penguins').dropna()
```

```python
sns.scatterplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', hue = 'species')
```

```python
sns.pairplot(penguins, hue = 'species')
```

```python
fig, ax = plt.subplots(1,2)
ax[0].hist(penguins['bill_length_mm'])
ax[1].boxplot(penguins['bill_length_mm'])
```

```python
fig, ax = plt.subplots(1,2)
sns.histplot(penguins['bill_length_mm'], ax = ax[0])
sns.boxplot(penguins['bill_length_mm'], ax = ax[1])
```
